## Student categorization and visualization 

```python
import random
```

```python
## generate a list of Nigerian names both male and female names

male_names = ["Adebayo","Akin", "chidi", "Collins", "Akoja","Ebenezar", "Emeka", "Femi", "Gbenga", "Jamal", "Yemi", "Nnamdi","Ibrahim", "Hassan", "Uche","Rafiu"]
female_names = ["Sola", "Bola", "Aisha", "Bisola", "Glorai", "Damilola", "Chinwe", " Tina", "Uju", "Jamila", "Zara", "Kafayat", "Precious", "Eunice", "Esosa","Wendy"]


```

```python
# print(len(male_names))
# print(len(female_names))

name = male_names + female_names
print(name)
```

```python
print(len(name))
```

```python
# Generating students names 

students = {
    "Male":[random.choice(male_names)
             for _ in range (16)],
    "Female": [random.choice(female_names)
               for _ in range(16)]
}

print(students)
```

# Catergorizing students into beautiful, handsdome, average, and normal

```python
def categorize_students(students):
    categories = ["Beautiful", "Handsome", "Average", "Normal"]
    categorize_students = {
        "Male": {
            category:[] for category in categories
        },
        "Female": {
            category: [] for category in categories
        }
    }
    for gender, names in students.items():
        for name in names:
            category = random.choice(categories)
            categorized_students[gender]
            [category].append(name)
            return categorized_students    
    categorized_students = categorize_students(students)
    print(categorized_students)
    
```

# Visualization 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# visualizing student categories 
# Data_set 
male_count = len(students["Male"])
female_count = len(students["Female"])
```

# Bar chart

```python
plt.figure(figsize= (8,5))
plt.bar(["Male","Female"],[male_count, female_count])
plt.title("Student Demography")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

# Pie Chart

```python
plt.figure(figsize= (8, 8))
plt.pie([male_count, female_count],labels=["Male", "Female"], autopct = "%1.1f%%")
plt.title("Students Demography")
plt.show()
```
